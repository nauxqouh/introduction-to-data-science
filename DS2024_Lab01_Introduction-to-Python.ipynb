```python
# Q8
def calc_sum(nums):
    sum = 0
    for i in nums:
      sum += i
    return sum

nums = [5, 3, -2, 4, -1]
calc_sum(nums)
```

```python
# Q9
def count_char_freq(input_string):
    char_freq = {} # emty dict
    for i in input_string:
      if i in char_freq:
        char_freq[i] += 1
      else:
        char_freq[i] = 1
    return char_freq

string = "abcaccdefeef"
char_freq = count_char_freq(string)
print(char_freq)
```
